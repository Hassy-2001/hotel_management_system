```python
# python assignment

import random

# collections
# we are going to use two collections here which include list and dictionary.

# this list will tell us the availability of room in hotel # True means Vacant and False means Occupied
room_availability = [True, True, True, False, False]

# this dictionary will tell is the room prices and its types
room_price_and_its_type = {"single": 50, "double": 70, "suite": 100}

# here we will store all the data of guest
complete_guest_information = {}


############################################# Start of Function section ############################################

# three user_defined_functions

# 01 function which is void and take one parameters
# with this function we will display room status

def room_status(room_number):
    status = "Vacant" if room_availability[room_number - 1] else "Occupied"
    print(f"{room_number} is currently {status}")


# 02 function which is fruitful function and don't take any parameter
# with this function we will generate booking reference number

def booking_reference():
    return f"REF{random.randint(1000, 9999)}"


# 03 function which is fruitful function and take two parameters
# with this function we will book a room in hotel

def room_booking(room_type, duration_of_stay):
    # first input
    room_number = int(input("Enter the room number (1-5): "))
    # using while loop
    while room_number < 1 or room_number > 5:
        print("Invalid room number!! Please enter the room number between 1 and 5")
        # same input again
        room_number = int(input("Enter the room number (1-5): "))

    room_status(room_number)
    # second input
    guest_name = input("Enter your name please: ")
    total_cost_of_stay = room_price_and_its_type.get(room_type, 0) * duration_of_stay

    # Use of nested condition for special offer
    if duration_of_stay >= 7:
        print("Special Offer !! 10% discount has been applied for staying 7 or more days in our hotel")
        total_cost_of_stay *= 0.9

    # updating data in dictionary
    complete_guest_information[booking_reference()] = {
        "Guest_name": guest_name, "Room_number": room_number,
        "Duration_of_stay": duration_of_stay, "Total_cost_of_stay": total_cost_of_stay}


# Anonymous Function which is Lambda function
convert_data = lambda x: x.upper()

########################################### End of Function Section  ###############################################

# Taking rest of three inputs from user

room_type_input = input("Enter the room type (single/double/suite): ")
duration_input = int(input("Enter the duration of stay (in days): "))
guest_name_input = input("Enter Your name please: ")

# Calling Function

room_status(2)
room_booking(room_type_input, duration_input)

# showing complete data of guest
print("\n Complete Guest Information: ")
for key, value in complete_guest_information.items():
    print(f"{key}:{value}")

    print("\n this is anonymous function\n")

# Use of anonymous function
transformed_name = convert_data(guest_name_input)
print(f"Guest name in UpperCase: {transformed_name}")

print("\n python collections")

# List advance Methods
print("\n List and its advance methods\n")

# it will append means increase the Vacant room in list
room_availability.append(True)
print("We have append the room availability: ", room_availability)

# it will reverse the order of list means from 1-7 to 7-1
room_availability.reverse()
print("We have reverse the room list: ", room_availability)

# it will extend means add two more availability in list
room_availability.extend([False, True])
print("We have extend the room availability: ", room_availability)

# it will remove  the one the first occupied room from the list
room_availability.remove(False)
print("We have remove the first occupied room: ", room_availability)

print("\n Dictionary and its advance methods\n")

# advance dictionary method

# it will add one more room type in hotel
room_price_and_its_type["pant house"] = 300
print("Adding one more room type in dictionary: ", room_price_and_its_type)

# it will remove  the last room type and its price
room_price_and_its_type.popitem()
print("removing last room type in dictionary: ", room_price_and_its_type)

# it will update the dictionary with new room type and its price
room_price_and_its_type.update({"Deluxe": 170, "Standard": 60})
print("Updating more room types in dictionary: ", room_price_and_its_type)

# it will clear the whole dictionary
room_price_and_its_type.clear()
print("it will clear the whole dictionary: ", room_price_and_its_type)


print("\n String and its advance methods\n")

# String and its four methods

Original_guest_name = "joe Root"
print("This is the original string: ", Original_guest_name)
# it will Capitalize the first letter
Capitalize = Original_guest_name.capitalize()
print("We have Capitalize the first letter of string: ", Capitalize)

# it will tell the length of name
Length = len(Original_guest_name)
print("We have the length of string: ", Length)

# it will check is the Original_guest_name is in alphabetic character
Isalpha = Original_guest_name.isalpha()
print("We are checking is our string is alphabetic characters : ", Isalpha)

# it will uppercase all the letter
Uppercase = Original_guest_name.upper()
print("We have uppercase the whole string: ", Uppercase)

# it will split the name into words list
Split = Original_guest_name.split()
print("We have been split string into word list: ", Split)


```
